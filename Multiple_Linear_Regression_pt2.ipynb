# 3. target variable analysis

```python
plt.subplots(figsize=(6,6)) #pyplot uses subplots function creates a new figure with specific width and height
sns.distplot(df['SalePrice'], fit=stats.norm) #seaborn calls distplot function which creates a histogram of dataframe saleprice, fit=stats.norm overlays a normal distibution from scipy.stats onto the sns plot
(mu, sigma) =  stats.norm.fit(df['SalePrice']) #the line calculates means and std which is stored in mu and sigma variable fit basically guesses the smoothest bell-shaped curve (normal distribution) that best describes the spread of your data.
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)], loc='best') #the line adds a legend to the plot, format is used to insert values of mu and std, loc='best' tells to choose the best location to legend
plt.ylabel('Frequency') # y label of the graph
plt.show() #show the plot
```

```python
fig = plt.figure()  #creates an empty  figure for plotting
stats.probplot(df['SalePrice'], plot=plt) #generates a probability plot of the salesprice column of df and assings the plot for customization
plt.show() #show the plot
```

## 3.1 This target varibale is right skewed. Now, we need to tranform this variable and make it normal distribution.  
  (Imagine someone strechtes right side of ruler a lot increasing the distance between high numbers this creates a right skew where lower numbers are are bunched up and high numbers are spread out)

### so we will use log function to make target variable more normal distribution

```python
df['SalesPrice'] = np.log1p(df['SalePrice']) #This code replaces the SalesPrice column with its natural logarithm after adding 1 to each value
```

```python
plt.subplots(figsize=(6,6)) #pyplot uses subplots function creates a new figure with specific width and height
sns.distplot(df['SalesPrice'], fit=stats.norm) #seaborn calls distplot function which creates a histogram of dataframe
(mu,sigma) = stats.norm.fit(df['SalePrice'])
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)], loc='best') #the line adds a legend to the plot
plt.ylabel('Frequency') # y label of the graph
plt.show() #show the plot
```

## 7. imputting missing values

```python
# PoolQC has missing value ratio is 99%+. So, there is fill by None
df['PoolQC'] = df['PoolQC'].fillna('None')
```

```python
#Arround 50% missing values attributes have been fill by None
df['MiscFeature'] = df['MiscFeature'].fillna('None')
df['Alley'] = df['Alley'].fillna('None')
df['Fence'] = df['Fence'].fillna('None')
df['FireplaceQu'] = df['FireplaceQu'].fillna('None')
```

```python
#Group by neighborhood and fill in missing value by the median LotFrontage of all the neighborhood
df['LotFrontage'] = df.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))
```

```python
#GarageType, GarageFinish, GarageQual and GarageCond these are replacing with None
for col in ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']:
    df[col] = df[col].fillna('None')
```

```python
#GarageYrBlt, GarageArea and GarageCars these are replacing with zero
for col in ['GarageYrBlt', 'GarageArea', 'GarageCars']:
    df[col] = df[col].fillna(int(0))
```

```python
#BsmtFinType2, BsmtExposure, BsmtFinType1, BsmtCond, BsmtQual these are replacing with None
for col in ('BsmtFinType2', 'BsmtExposure', 'BsmtFinType1', 'BsmtCond', 'BsmtQual'):
    df[col] = df[col].fillna('None')
```

```python
#MasVnrArea : replace with zero
df['MasVnrArea'] = df['MasVnrArea'].fillna(int(0))
```

```python
#MasVnrType : replace with None
df['MasVnrType'] = df['MasVnrType'].fillna('None')
```

```python
#There is put mode value
df['Electrical'] = df['Electrical'].fillna(df['Electrical']).mode()[0]
```

```python
#There is no need of Utilities
df = df.drop(['Utilities'], axis=1)
```

```python
#Checking there is any null value or not
plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull())
```

## 8. no missing values encoding str to int

```python
cols = ('FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond',
        'ExterQual', 'ExterCond','HeatingQC', 'PoolQC', 'KitchenQual', 'BsmtFinType1',
        'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 'LandSlope',
        'LotShape', 'PavedDrive', 'Street', 'Alley', 'CentralAir', 'MSSubClass', 'OverallCond',
        'YrSold', 'MoSold', 'MSZoning', 'LandContour', 'LotConfig', 'Neighborhood',
        'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
        'Exterior2nd', 'MasVnrType', 'MasVnrArea', 'Foundation', 'GarageType', 'MiscFeature',
        'SaleType', 'SaleCondition', 'Electrical', 'Heating')
```

```python
from sklearn.preprocessing import LabelEncoder
for c in cols:
    lbl = LabelEncoder()
    lbl.fit(list(df[c].values))
    df[c] = lbl.transform(list(df[c].values))
```

## 9 data for prediction

```python
y = df['SalesPrice']
X = df.drop(['SalesPrice'], axis=1)
```

```python
X = X.values
y = y.values
```

```python
# Split data into train and test formate
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

## 10. Linear regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#Fit the model
model.fit(X_train, y_train)
```

```python
#Prediction
print("Predict value " + str(model.predict([X_test[142]])))
print("Real value " + str(y_test[142]))
```

```python
#Score/Accuracy
print("Accuracy --> ", model.score(X_test, y_test)*100)
```
